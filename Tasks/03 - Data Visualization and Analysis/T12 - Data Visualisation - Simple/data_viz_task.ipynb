# Data visualisation task 

**Additional information**  
[Cars93 metadata](https://jse.amstat.org/v1n1/datasets.lock.html) 

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv("Cars93.csv", index_col=0, keep_default_na=False)
```

### 1. Boxplot

```python
#Filter the car data
manufacturers_box = ["Audi", "Hyundai", "Suzuki", "Toyota"]
car_data_box = car_data[car_data["Manufacturer"].isin(manufacturers_box)]
# Boxplot for revs per mile
sns.boxplot(data=car_data_box, x="Manufacturer", y="Rev.per.mile")
plt.title("Rev Per Mile and Manufacturer")
plt.xlabel("Manufacturer")
plt.ylabel("Rev Per Mile")
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**
Toyota


### 2. Histogram

```python
# Histogram of MPG
sns.histplot(car_data["MPG.city"], bins=20, kde=True, color="blue")
sns.histplot(car_data["MPG.highway"], bins=20, kde=True, color="orange")
plt.title("Histogram of MPG City(Blue) and MPG Highway(Orange)")
plt.xlabel("Miles per Gallon (MPG)")
plt.ylabel("Frequency")
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
On the highway.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(data=car_data, x="Wheelbase", y="Turn.circle")
plt.title("Relationship between Wheelbase and Turning circle")
plt.xlabel("Wheelbase")
plt.ylabel("Turning circle")
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
Positive Correlation, when the wheelbase gets larger, the turning circle increase.


### 4. Barplot

```python
# Horsepower mean for eache car type
car_data_group = car_data.groupby("Type")["Horsepower"].mean().reset_index()
# Barplot for mean horsepower
sns.barplot(data=car_data_group, x="Type", y="Horsepower")
plt.title("Mean Horsepower and Car Type")
plt.xlabel("Car Type")
plt.ylabel("Mean Horsepower")
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes
