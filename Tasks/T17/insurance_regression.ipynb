## How age affects insurance costs?

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the insurance data
insurance_df = pd.read_csv("insurance.csv")
insurance_df.head()
```

```python
# select age and charges rows from the insurance dataframe
x = insurance_df.loc[:, "age"].values
y = insurance_df.loc[:, "charges"].values

# Reshape the data
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Plot the scatter plot
plt.scatter(x=x, y=y, color="b")
plt.xlabel("Age")
plt.ylabel("Insurance Costs")
plt.title("Scatter Plot")
plt.show()
```

```python
# fit a model
charges_model = LinearRegression()
charges_model.fit(X=x, y=y)
y_predict = charges_model.predict(X=x)

# Plot the scatter plot and best-fit line
plt.scatter(x=x, y=y, color="b")
plt.plot(x, y_predict, color="r", label="Best-fit Line")
plt.xlabel("Age")
plt.ylabel("Insurance Costs")
plt.title("Scatter Plot and Best-fit Line")
plt.legend()
plt.show()
```
