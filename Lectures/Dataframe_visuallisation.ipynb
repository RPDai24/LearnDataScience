```python
#pip install numpy pandas matplotlib seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Check available dataset
#sns.get_dataset_names()

#Built in titanic dataset
df = sns.load_dataset("titanic")

# #Fetch data from Github
# url = "https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv"
# df = pd.read_csv(url)
# #Read cse file fro local dir 
# filename = "titanic.cvs"
# df = pd.read_csv(url)

# df.info()

df.describe()

```

```python
df.columns
```

```python
#Drop null values
new_df = df.dropna()
new_df.info()
```

```python
#Replace null value with mean, median or mode
x = df["age"].mean()
df["age"].fillna(x, inplace = True) # inplace = True overwrites the dataframe
```

```python
df.info()
```

Visualisation

```python
sns.histplot(df["age"], bins=30, color="g", kde=True) # Kernel density 
plt.savefig("hist.png")
```

```python
#Removing rows
for x in df.index:
    if df.loc[x, "age"] > 50:
        df.drop(x, inplace=True)
sns.histplot(df["age"], bins=30, color="y", kde=True)  
```

```python
sns.kdeplot(df["age"], shade=True) 
```

```python
sns.countplot(data=df, x="survived", palette="Set1")
```
